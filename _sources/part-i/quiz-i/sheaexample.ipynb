```python
import json
from base64 import b64encode, b64decode
from jupyterquiz import display_quiz
from data import query
```

```python
example=[{
        "question": "The variable mylist is a Python list. Choose which code snippet will append the item 3 to mylist.",
        "type": "multiple_choice",
        "answers": [
            {
                "code": "mylist+=3",
                "correct": False
            },
            {
                "code": "mylist+=[3]",
                "correct": True
            },
            {
                "code": "mylist+={3}",
                "correct": False
            }
        ]
    }]
```

```python
display_quiz(query)
```

```python
display_quiz(example)
```

```python
json.dumps(example)
```

```python
display_quiz("#question1")
```

```python
encoding = b64encode(bytes(json.dumps(example), 'utf8'))
encoding
```

```python
decoded = encoding.decode('utf-8')
bit_decoded = b64decode(decoded)
utfd = bit_decoded.decode('ascii')
json_str=json.loads(utfd)
json_str
display_quiz(json_str)
```
